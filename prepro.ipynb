# Procesamiento de imagenes deigitales

una imagen digital es una matriz de pixeles o puntos de muestreo qye almacena información sobre la intensidad de color en la ubicación de cada pixel.

### RGB
RGB es un modelo cromático mediante el cual podemos representar distintos colores a partir de la mezcla de estos tres colores primarios.

Cada uno tiene muestra un discreto entre 0 y 255.

A una imagen se le puede:
- Modificar
- Ajustar la escala
- Rotar
- Cambiar el color
- Cambiar brillo

Imagen binaria es a blanco y negro, a color es un cubo.






```python
import cv2
from google.colab.patches import cv2_imshow

# Cargar la imagen
imagen1 = cv2.imread('montypython.jpg')

if imagen1 is None:
    print("Error: La imagen no se pudo cargar. Verifica la ruta o el archivo.")
else:
    # Mostrar la imagen en Google Colab
    cv2_imshow(imagen1)
```

```python
print(imagen1)
```

```python
import matplotlib.pyplot as plt

if imagen1 is None:
    print("Error: La imagen no se pudo cargar. Verifica la ruta o el archivo.")
else:
    # Convertir la imagen de BGR (usado por OpenCV) a RGB (usado por Matplotlib)
    imagen_rgb = cv2.cvtColor(imagen1, cv2.COLOR_BGR2RGB)

    # Crear la figura
    plt.figure(figsize=(8, 6))

    # Mostrar la imagen
    plt.imshow(imagen_rgb)
    plt.axis('on')  # Puedes usar 'off' para ocultar los ejes
    plt.title("Imagen cargada como matriz")
    plt.show()
```

```python
plt.figure(figsize=(8, 6))

# Mostrar la matriz de la imagen como mapa de calor
plt.imshow(imagen1, cmap='viridis')  # Prueba otros cmap como 'gray', 'hot', 'cool', etc.
plt.colorbar(label="Intensidad de píxeles")
plt.title("Imagen como matriz (Grayscale)")
plt.show()
```
